```r
init_data_work <- function(data){
  data.work <- data
  data.work$target <- as.factor(data.work$Readmission.Status)
  data.work$Race <- as.factor(data.work$Race)
  data.work$DRG.Class <- as.factor(data.work$DRG.Class)
  data.work$DRG.Complication <- as.factor(data.work$DRG.Complication)
  data.work <- data.work[, !(names(data.work) %in% list('Readmission.Status'))]
  return(data.work)
}
data.work <- init_data_work(data)
head(data.work)
summary(data.work)
```

```r
count_rate <- function(df, col_name, col_values){
  res <- list()
  for(col_value in col_values){
    res[col_value] <- (nrow(df[(df[col_name]==col_value) & (df['target']==1), ]) / 
                      nrow(df[(df[col_name]==col_value), ]))
  }
  return(res)
}
```

```r
gender_mapper <- function(gender) {
  return(ifelse(gender == 'M', 2,
         ifelse(gender == 'F', 1, 0)))
}
race_mapper <- function(race) {
  return(ifelse(race == 'Hispanic', 4,
         ifelse(race == 'Black', 3,
         ifelse(race == 'White', 2,
         ifelse(race == 'Others', 1, 0)))))
}
drg_mapper <- function(drg){
  return(ifelse(drg == 'UNGROUP', 3,
         ifelse(drg == 'MED', 2,
         ifelse(drg == 'SURG', 1, 0))))
}
comp_mapper <- function(comp){
  return(ifelse(comp=='Others', 5,
         ifelse(comp=='MedicalMCC.CC', 4,
         ifelse(comp=='MedicalNoC.CC', 3,
         ifelse(comp=='SurgMCC.CC', 2,
         ifelse(comp=='SurgNoC', 1, 0))))))
}
```

```r
one_hot <- function(df, old_col, values) {
  for(value in values){
    df[, value] <- ifelse(df[, old_col]==value, 1, 0)
  }
  return(df)
}
set_col_to_category <- function(df, old_col, new_col, mapper){
  df[, new_col] <- as.factor(mapper(df[, old_col]))
  return(df)
}
set_col_to_numerical_factor <- function(df, old_col, new_col) {
  df[, new_col] <- as.factor(as.numeric(as.factor(df[, old_col])))
  return(df)
}
set_col_to_log <- function(df, old_col, new_col){
  df[, new_col] <- df[, old_col]
  df[df[new_col]==0, new_col] <- 0.01
  df[, new_col] <- log10(df[, new_col])
  return(df)
}
set_col_to_scale <- function(df, old_col, new_col){
  # df[, new_col] <- scale(as.numeric(df[, old_col]))
  df[, new_col] <- scale(sapply(df[, old_col], as.numeric))
  return(df)
}
remove_cols <- function(df, ...){
  cols <- unlist(list(...))
  
  new_df <- df[, !(names(df) %in% cols)]
  
  return(new_df)
}
```

```r
wrangling_process_1 <- function(df) {
  df <- set_col_to_category(df, 'Gender', 'Gender_1', gender_mapper)
  df <- set_col_to_category(df, 'Race', 'Race_1', race_mapper)
  df <- set_col_to_category(df, 'DRG.Class', 'DRG.Class_1', drg_mapper)
  df <- set_col_to_category(df, 'DRG.Complication', 'DRG.Complication_1', comp_mapper)
  df <- remove_cols(df, 'Gender', 'Race', 'DRG.Class', 'DRG.Complication')
  return(df)
}
wrangling_process_2 <- function(df) {
  df <- one_hot(df, 'Gender', gender_values)
  df <- one_hot(df, 'Race', race_values)
  df <- one_hot(df, 'DRG.Class', drg_values)
  df <- one_hot(df, 'DRG.Complication', comp_values)
  df <- set_col_to_scale(df, 'HCC.Riskscore', 'HCC.Riskscore_1')
  df <- set_col_to_scale(df, 'Age', 'Age_1')
  df <- set_col_to_scale(df, 'LOS', 'LOS_1')
  df <- set_col_to_scale(df, 'ER', 'ER_1')
  df <- remove_cols(df, 'Gender', 'Race', 'DRG.Class', 'DRG.Complication', 'Age', 'HCC.Riskscore', 'LOS', 'ER')
  return(df)
}

wrangler <- wrangling_process_2

data.procd <- wrangler(data.work)
head(data.procd)

# data.fin <- SMOTE(target ~ ., data.procd, perc.over = 200, perc.under=200)
# head(data.fin)
# summary(data.fin)
```

```r
build_fold_indices <- function(fold) {
  test_fold <- 1:fold
  shifted <- (test_fold + 1)
  valid_fold <- ifelse((shifted > fold), shifted %% fold, shifted)
  fold_indices <- data.frame(test = test_fold, valid = valid_fold)
  return(fold_indices)
}
pick_out_indices <- function(total_rows, fold, ...) {
  indices <- c()
  for (i in list(...)) {
    indices <- append(indices, seq(i, total_rows, fold))
  }
  indices <- unique(indices)
  return(indices)
}
get_by_exclude <- function(df, fold, ...) {
  total_rows <- nrow(df)
  exclude <- pick_out_indices(total_rows, fold, ...)
  excluded <- df[-exclude,]
  return(excluded)
}
get_by_index <- function(df, fold, ...) {
  total_rows <- nrow(df)
  include <- pick_out_indices(total_rows, fold, ...)
  included <- df[include,]
  return(included)
}
get_cols_in_df <- function(df, ...){
  cols <- unlist(list(...))
  
  new_df <- df[, (names(df) %in% cols)]
  
  return (new_df)
}
shuffle_frame <- function(df){
  set.seed(28657)
  shf_rows <- sample(nrow(df))
  shf_train <- df[shf_rows, ]  
  return(shf_train)
}
subset_by_fold <- function(df, fold, index, indices){
  training_frame <- get_by_exclude(df,
                                   fold,
                                   indices[index, ]$test,
                                   indices[index, ]$valid)
  validation_frame <- get_by_index(df, fold, indices[index, ]$valid)
  testing_frame <- get_by_index(df, fold, indices[index, ]$test)
  
  return(list(training=training_frame,
              validation=validation_frame,
              testing=testing_frame))
}
build_cm <- function(df, pred) {
  result_frame <- data.frame(truth = df$target,
                             pred = pred)
  result_table <- table(result_frame)
  return(result_table)
}
cal_pred_result <- function(cM) {
  true_col <- 2
  false_col <- 1
  positive_row <- 2
  negative_row <- 1
  
  TP <- cM[positive_row, ][true_col]
  FN <- cM[positive_row, ][false_col]
  TN <- cM[negative_row, ][false_col]
  FP <- cM[negative_row, ][true_col]
  
  result <- c(TP, FP, TN, FN)
  names(result) = c("TP", "FP", "TN", "FN")
  return(result)
}
get_rounded <- function(value, round_digits = 2) {
  return(round(value, digits = round_digits))
}
cal_accu <- function(pred_result) {
  TP <- unname(pred_result["TP"])
  FN <- unname(pred_result["FN"])
  TN <- unname(pred_result["TN"])
  FP <- unname(pred_result["FP"])
  total_rows = TP + FN + TN + FP
  accuracy <- (TP + TN) / total_rows
  return(get_rounded(accuracy))
}
cal_sensitivity <- function(pred_result) {
  TP <- unname(pred_result["TP"])
  FN <- unname(pred_result["FN"])
  sensitivity <- TP / (TP + FN)
  return(get_rounded(sensitivity))
}
cal_specificity <- function(pred_result) {
  TN <- unname(pred_result["TN"])
  FP <- unname(pred_result["FP"])
  specificity <- TN / (TN + FP)
  return(get_rounded(specificity))
}
cal_F1 <- function(pred_result) {
  precision <- cal_precision(pred_result)
  recall <- cal_sensitivity(pred_result)
  F1 <- (2 * precision * recall) / (precision + recall)
  return(get_rounded(F1))
}
cal_precision <- function(pred_result) {
  TP <- unname(pred_result["TP"])
  FP <- unname(pred_result["FP"])
  precision <- TP / (TP + FP)
  return(get_rounded(precision))
}
evaluate_model <- function(df, pred, threshold=0.5){
  pred_class <- ifelse(pred[, 2] > pred[, 1], 1, 0)
  pred_prob <- pred
  
  predict_cm <- build_cm(df, pred_class)
  predict_result <- cal_pred_result(predict_cm)
  accu <- cal_accu(predict_result)
  preci <- cal_precision(predict_result)
  sens <- cal_sensitivity(predict_result)
  recl <- sens
  spec <- cal_specificity(predict_result)
  f1 <- cal_F1(predict_result)
  
  # svm
  # attr(pred_prob, 'probabilities')
  # roc_pred <- prediction(attr(pred_prob, 'probabilities')[, 1], df$target)
  
  # normal
  roc_pred <- prediction(pred_prob[, 2], df$target)
  
  roc_perf <- performance(roc_pred, measure = 'auc')
  r2 <- cor(as.numeric(df$target), as.numeric(pred_class)) ^ 2
  
  return(list(cm=predict_cm,
              result=predict_result,
              accuracy=accu,
              recall=recl,
              precision=preci,
              sensitivity=sens,
              specificity=spec,
              F1=f1,
              AUC=roc_perf@y.values[[1]],
              R2=r2))
}
```

```r
# FORMULA <- 'target ~ ER + LOS + Age + HCC.Riskscore + Gender_1 + Race_1 + DRG.Class_1 + DRG.Complication_1'
FORMULA <- 'target ~ .'
```

```r
train_dt <- function(df, depth=5) {
  model <- rpart(
    as.formula(FORMULA),
    data = df,
    control = rpart.control(maxdepth = depth),
    method = "class"
  )
  return (model)
}
predict_dt <- function(model, df) {
  pred <- predict(model, df, type = "prob")
  return (pred)
}
```

```r
train_rf <- function(df, mtry=3, ntree=500){
  model <- randomForest(
    as.formula(FORMULA),
    data=df,
    mtry = mtry,
    ntree = ntree,
    importance=TRUE,
    proximity=TRUE)
  return(model)
}
predict_rf <- function(model, df){
  pred <- predict(model, df, type='prob')
  return(pred)
}
```

```r
train_sv <- function(df, kernel='linear') {
  model <- svm(
    as.formula(FORMULA),
    data = df,
    kernel = kernel,
    probability = TRUE)
  return(model)
}
predict_sv <- function(model, df) {
  pred <- predict(model, df, probability = TRUE)
  return(pred)
}
```

```r
train_rf_r <- function(df, mtry=3, ntree=500, importance='none', node_size=3, sample_size=0.632){
  model <- ranger::ranger(formula = as.formula(FORMULA),
                          data=df,
                          num.trees = ntree,
                          mtry = mtry,
                          importance = importance,
                          min.node.size = node_size,
                          sample.fraction = sample_size,
                          num.threads = 4,
                          save.memory = TRUE,
                          keep.inbag = TRUE,
                          probability = TRUE)
  return(model)
}
predict_rf_r <- function(model, df){
  pred <- predict(model, df, type='se')
  return(pred)
}
```

```r
# reference: https://www.jamleecute.com/random-forests-%E9%9A%A8%E6%A9%9F%E6%A3%AE%E6%9E%97/

hyper_grid <- expand.grid(
  mtry=seq(3,7),
  node_size=seq(3,9, by=2),
  sample_size = c(0.55, 0.632, 0.7, 0.8),
  importance = c('none', 'impurity', 'impurity_corrected', 'permutation'),
  OOB_RMSE=0)

nrow(hyper_grid)
head(hyper_grid)
```

```r
for (i in 1:nrow(hyper_grid)) {
  model <- ranger::ranger(
    formula = as.formula(FORMULA),
    data=splitted_frames$training,
    num.trees = 500,
    importance = hyper_grid$importance[i],
    mtry = hyper_grid$mtry[i],
    min.node.size = hyper_grid$node_size[i],
    sample.fraction = hyper_grid$sample_size[i],
    num.threads = 4,
    save.memory = TRUE,
    keep.inbag = TRUE,
    probability = TRUE)
 
  hyper_grid$OOB_RMSE[i] <- sqrt(model$prediction.error)
}
 
hyper_grid %>% 
  dplyr::arrange(OOB_RMSE) %>% 
  head(10)
```
